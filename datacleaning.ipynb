```python
import numpy as np
import pandas as pd

#loading dataset
df = pd.read_csv("cafedataset.csv")


df.head()
```

```python
#repalcing error and unknown values with nan beacuse python understand nan values
df.replace(["ERROR", "UNKNOWN"], np.nan, inplace=True)


#converting to data formats like "2"->2 (str->int values)
df['Quantity']=pd.to_numeric(df['Quantity'],errors='coerce')
df['Price Per Unit']=pd.to_numeric(df['Price Per Unit'],errors='coerce')
df['Total Spent']=pd.to_numeric(df['Total Spent'],errors='coerce')
df['Transaction Date']=pd.to_datetime(df['Transaction Date'],errors='coerce')

#handling missing values
df['Item'].fillna(df['Item'].mode()[0])
df['Price Per Unit'].fillna(df["Price Per Unit"].mean())
df['Total Spent'].fillna(df['Total Spent'].mean())
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']
df['Quantity'].fillna(df['Quantity'].median())
df['Payment Method'].fillna(df['Payment Method'].mode()[0])
df['Location'].fillna(df['Location'].mode()[0])
df['Transaction Date'].fillna(df['Transaction Date'].mode()[0])


#calculating Total Bill
df['Calculated Total'] = df['Quantity'] * df['Price Per Unit']
df['Day'] = df['Transaction Date'].dt.day_name()


print("Cleaned dataset saved successfully!")
df.head()
```

```python
df.to_csv('cleaned_cafe_sale.csv',index=False)
```
